```python
'''
LSTM for sequence replication
'''

import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras import *
import matplotlib.pyplot as plt
import tensorflow as tf
print('done')
```

```python
fig=plt.gcf()
fig.canvas.draw()

x=np.array([i for i in range(0,100)])
data = np.zeros((1,100, 1))
data[0,30:,0]=1.0
target= np.zeros((1,100,1))
for i in range(100):
    target[0,i,0] = data[0,i,0]*((1-np.exp(-(i-30)/5.0)))

for t in range(8):
    plt.gcf().clear()

    #plt.plot(data[0,0,:])
    plt.plot(target[0,:100,0])
    plt.plot(data[0,:100,0])
    fig.canvas.draw()
    plt.pause(0.001)

```

```python

x_test = np.zeros((1,100,1))
x_test[0,50:,0]=1.0
y_test = np.zeros((1,100,1))
for i in range(100):
    y_test[0,i,0] = x_test[0,i,0]*((1-np.exp(-(i-50)/5.0)))



model = Sequential()  
model.add(LSTM(6, return_sequences=True, input_shape=(None, 1)))
model.add(LSTM(6,return_sequences=True))
model.add(Dense(1))
model.compile(loss='mse', optimizer='rmsprop',metrics=['accuracy'])
model._make_predict_function()

default_graph = tf.get_default_graph()
try:
    model.load_weights("Models/lstmmin_size_corrected.model")
    print('weights loaded ')
except:
    print('weights not loaded ')

print('done')
```

```python
x = tf.placeholder(tf.float32, shape=(None,100,1))
y = tf.placeholder(tf.float32, shape=(None,100,1))

p,v=model(x)

```

```python
for i in range(10):

    model.fit(data, target, epochs=10, batch_size=1, verbose=2,validation_data=(x_test, y_test))
    predict = model.predict(x_test)
    plt.gcf().clear()
    predict = model.predict(data)
    plt.plot(x,predict[0,:,0])
    plt.plot(x,target[0,:,0])
    fig.canvas.draw()
    plt.pause(0.01)
    #plt.show()
print('done')
```

```python
def check_predict(inp,exp_out):
    predict = model.predict(inp)
    plt.gcf().clear()
    plt.plot(predict[0,:,0])
    plt.plot(inp[0,:,0])
    plt.plot(exp_out[0,:,0])
    plt.pause(0.01)
```

```python
def generate_seq(t):
    x_test = np.zeros((1, 1, 200))
    x_test[0,0,t:]=1.0
    y_test = np.zeros((1, 1, 200))
    for i in range(200):
        y_test[0,0,i] = x_test[0,0,i]*((1-np.exp(-(i-t)/5.0)))
    return x_test.reshape((1,200,1)),y_test.reshape((1,200,1))
```

```python
for i in range(20):
    xt,y=generate_seq(10*i+3)
    check_predict(xt,y)
```
